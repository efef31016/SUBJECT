# Problem_CG

```python
length = 40

# row
A=[]

for i in range(length):
    # col
    A_temp=[]
    for j in range(length):
        if i==j:
            A_temp.append(2*(i+1))
        elif j==i+1 or j==i-1:
            A_temp.append(-1)
        else:
            A_temp.append(0)
        
    A.append(A_temp)
b = [1.5*(i+1)-6 for i in range(length)]

x_init = [0 for i in range(length)]
```

```python
def CG(A, b, x_init, tol=1e-5):
    import numpy as np
    
    xk = x_init
    rk = b - np.dot(A, xk)
    vk = rk
    rk_norm = np.linalg.norm(rk)
    
    num_iter = 0
    x_record = [xk]
    r_record = [rk]
    while rk_norm > tol:
        avk = np.dot(A, vk)
        rkrk = np.dot(rk, rk)
        
        tk = rkrk / np.dot(vk, avk)
        xk = xk + tk * vk
        rk = rk - tk * avk
        sk = np.dot(rk, rk) / rkrk
        vk = rk + sk * vk
        
        num_iter += 1
        x_record.append(xk)
        r_record.append(rk)
        rk_norm = np.linalg.norm(rk)
    
    return np.array(x_record), np.array(r_record), x_record[num_iter]
```

### We print out the change of approximated solution each iteration.

```python
hist_x, hist_r, sol = CG(A,b,x_init)
```

```python
for i in hist_x[1:]:
    print('\nSolution: \t x = {}'.format(i))
```

### We print out the change of resisual each iteration.

```python
for i in hist_r[1:]:
    print('\nResisual: \t x = {}'.format(i))
```
